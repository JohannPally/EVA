```python
import serial as ser
import pandas as pd
import numpy as np
import time

PORT = "/dev/cu.usbmodem14203"

all_data = []
```

```python
my_ser_port = ser.Serial(PORT, 115200, timeout=0.1)
serial_cmd = "r"
encoded_cmd = serial_cmd.encode()

encoded_cmd
```

```python
def request_data():
    my_ser_port.write(encoded_cmd)
    
    serial_obj = my_ser_port.readline()
    # Poll the serial port
    line = str(serial_obj, 'utf-8')
    
    return line
```

```python
serialport = ser.Serial(PORT, 115200, timeout=0.1)

def get_data():
    global serialport
    
#         serialport.readline()
    serial_obj = serialport.readline()
    # Poll the serial port
    line = str(serial_obj, 'utf-8')
    ls = line.replace("\r\n", "").split("' ")
    
    if (len(ls) != 15):
        serial_obj = serialport.readline()
        # Poll the serial port
        line = str(serial_obj, 'utf-8')
        ls = line.replace("\r\n", "").split("' ")
    return np.array(ls, dtype=float)
    print(ls)
```

```python
np_array_all = np.array([])
df_to_convert = pd.DataFrame({"x_xl": [0.0], "y_xl": [0.0], "z_xl": [0.0], 
                              "x_gr": [0.0], "y_gr": [0.0], "z_gr": [0.0], 
                              "x_mg": [0.0], "y_mg": [0.0], "z_mg": [0.0], 
                              "orient_roll": [0.0], "orient_pitch": [0.0], "orient_yaw": [0.0], 
                              "pos_x": [0.0], "pos_y": [0.0], "pos_z": [0.0],
                              "timestamp": [0]})
# df_to_convert.columns = ["x_xl", "y_xl", "z_xl", "x_gr", "y_gr", "z_gr", "x_mg", "y_mg", "z_mg", "orient_roll", "orient_pitch", "orient_yaw", "pos_x", "pos_y", "pos_z"]
```

```python
def add_rows_to_df():
    global np_array_all
    #Time in seconds since the epoch
    time_stamp = time.time_ns()
    ls = get_data()
    ls = np.append(ls, time_stamp)
    
    if len(np_array_all) == 0:
        np_array_all = np.array(ls)
        df_to_convert.loc[len(df_to_convert.index)] = ls
        df_to_convert.loc[len(df_to_convert.index)-1]["timestamp"] = time_stamp

        
    else:
        np_array_all = np.vstack((np_array_all, ls))
        df_to_convert.loc[len(df_to_convert.index)] = ls
        df_to_convert.loc[len(df_to_convert.index)-1]["timestamp"] = time_stamp


    
```

```python
add_rows_to_df()
```

```python
def loop_func():
    while(1):
        add_rows_to_df()
```
